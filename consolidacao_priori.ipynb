```python
import pandas as pd
cbja_priori = pd.read_excel("PLANILHA CBJA PRIORI.xlsx")
# display(cbja_priori)
cbja_priori = cbja_priori.rename(columns={"ATIVO":"tombo", "VALOR JUSTO": "valor_priori"})
print(cbja_priori.info())
```

```python
cbja_suap = pd.read_excel("2024-07-31_10-34_cbja_suap_relatorio_patrimonio.xlsx")
cbja_suap["valor"] = cbja_suap["valor"].str.replace(".", "")
cbja_suap["valor"] = cbja_suap["valor"].str.replace(",", ".")
# display(cbja_suap)
cbja_suap["valor"] = pd.to_numeric(cbja_suap["valor"])
cbja_suap = cbja_suap.rename(columns={"valor":"valor_suap"})
print(cbja_suap.info())
```

```python
consolidacao_cbja = pd.merge(cbja_priori, cbja_suap, on="tombo", how="outer", indicator=True)
display(consolidacao_cbja)
```

```python
consolidacao_cbja.to_excel("consolidacao_cbja.xlsx", index=False)
```
